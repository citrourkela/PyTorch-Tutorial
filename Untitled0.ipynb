```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Model(nn.Module):
  def __init__(self, in_feature=4, h1=8, h2=9, out_feature=3):
    super().__init__()
    self.fc1 = nn.Linear(in_feature, h1)
    self.fc2 = nn.Linear(h1, h2)
    self.out= nn.Linear(h2, out_feature)

  def forward(self, x):
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.out(x)
    return x
torch.manual_seed(41)
model = Model()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url = 'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
my_df = pd.read_csv(url)
my_df['variety'] = my_df['variety'].replace('Setosa', 0.0)
my_df['variety'] = my_df['variety'].replace('Versicolor', 1.0)
my_df['variety'] = my_df['variety'].replace('Virginica', 2.0)
```

```python
# Trian, Test & Split
X = my_df.drop('variety', axis=1)
y = my_df['variety']
# Convert these to numpy array
X = X.values
y = y.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
#convert X & y to tensor
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
# To test the model error and devoiation of error
criterion = nn.CrossEntropyLoss()
# choose optimizer(Adam optimizer)
# lr is the learning rate(if error doesnot go downafter a bounch of iteration(epochs) then lower the learning rate)
# If loss is not decreasing â†’ try smaller rate like 0.001
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
# Train our model
epochs = 100 # epochs is one run through all the training data in our network
losses= []
for i in range (epochs):
  # Go forward to get the prediction
  y_pred = model.forward(X_train)
  # measure the loss/error
  loss = criterion(y_pred, y_train)
  # keep track of our losses/errors
  losses.append(loss.detach().numpy())
  # Print every 10 epochs
  if i % 10 == 0:
    print(f'Epoch: {i} and loss: {loss}')
  # Do some back propagation: take the error rate of forward propagation and feed it back through the network to fine tune the weight
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
# Graph it out
plt.plot(range(epochs), losses)
plt.ylabel("Loss/Error")
plt.xlabel("Epochs")
```

```python
with torch.no_grad():
  y_eval = model.forward(X_test)
  loss = criterion(y_eval, y_test)
```
